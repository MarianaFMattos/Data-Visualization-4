# Espaço de desafios

Notebook para o desenvolvimento de códigos de desafio do curso Data Visualization: estilização de tabelas com Python.

### Dicionário de variáveis

* **pais**: país de destino do produto.
* **id_cliente**: código de identificação do cliente solicitante.
* **data_pedido**: data de solicitação do produto por parte do cliente.
* **data_chegada**: data de chegada do produto ao endereço de entrega informado pelo cliente.
* **tipo_compra**: categoria que se enquadra a compra, podendo ser *Normal* ou *Devolução*.
* **numero_pedido**: código de identificação da compra realizada.
* **tamanho_pacote**: tamanho do pacote para envio do pedido.
* **unidades**: quantidade de unidades compradas no pedido.
* **custo_empacotamento**: valor do custo da loja para empacotar o produto.
* **custo_envio**: valor do custo da loja para enviar o produto.
* **tipo_desconto**: categoria do tipo de desconto foi aplicado.
* **categoria**: categoria de tipo de produto.
* **tipo_consumo**: nível de consumo do produto por tempo de uso.
* **tipo_cliente**: categoria de tipo de cliente.
* **custo_produto**: valor total do custo que o pedido resulta para a loja.

## Hora da prática 1

```python
import pandas as pd
df = pd.read_csv('loja_livro_filmes.csv')
df
```

```python
df.info()
```

```python
df['data_pedido'] = pd.to_datetime(df['data_pedido'])
df['data_chegada'] = pd.to_datetime(df['data_chegada'])
df.info()
```

```python
df_paises = df.groupby(['pais'])['unidades'].sum().nlargest(10).copy()
df_paises = df_paises.reset_index()
df_paises
```

```python
df_paises.columns = ['Países', 'Unidades Pedidas']
df_paises['Ranque'] = df_paises.index + 1
df_paises.set_index('Ranque', inplace = True)
df_paises
```

```python
s_paises = df_paises.style
s_paises = s_paises.format({'Unidades pedidas': '{} prod'})
s_paises
```

## Hora da prática 2

```python
df_custo_categoria = df.groupby(['categoria'])[['custo_produto']].sum()
df_custo_categoria.index.name = 'Categoria Produto'
df_custo_categoria.columns = ['Custo de pedidos']
df_custo_categoria
```

```python
estilo_categoria = df_custo_categoria.style
estilo_categoria.format('R$ {:,.2f}').background_gradient('Greens')
estilo_categoria
```

```python
estilo_categoria.format('R$ {:,.2f}')\
                .highlight_max(color = '#F6165')\
                .highlight_min(color = 'lightgreen')
```

## Hora da prática 3

```python
df_tipo_desconto = pd.DataFrame(df['tipo_desconto'].value_counts())
df_tipo_desconto.columns = ['N pedidos']
df_tipo_desconto.index.name = 'Descontos'
df_tipo_desconto
```

```python
import numpy as np
porcentagem = df_tipo_desconto['N pedidos'].to_numpy()
porcentagem = 100 * porcentagem / porcentagem.sum()
porcentagem
```

```python
df_tipo_desconto['Porcentagem por tipo de desconto'] = porcentagem
df_tipo_desconto
```

```python
estilo_desconto = df_tipo_desconto.style
estilo_desconto

cabecalho = {
    'selector': 'th',
    'props': 'font-weight: bold; font-family: Arial; text-align: center; background-color: white'
}
celulas = {
    'selector': 'td',
    'props': 'font-family: Arial; text-align: left; background-color: white'
}
estilo_desconto.set_table_styles([cabecalho, celulas], overwrite = False)
```

```python
estilo_desconto.format({'Porcentagem por tipo de desconto': '{:.2f} %'})\
               .bar(subset = 'Porcentagem por tipo de desconto', vmin = 0, vmax = 100.0, color = '#9CD33B',
                   height = 50, width = 60)
```

## Hora da prática 4

```python
df_meses = df.copy()
df_meses = df_meses.sort_values('data_pedido')
df_meses = df_meses.reset_index(drop = True)
df_meses = df_meses[df_meses['data_pedido'] >= '2013-01-01']
df_meses['meses'] = df_meses['data_pedido'].dt.strftime('%Y - %b')
df_meses
```

```python
tempo_demora = df_meses['data_chegada'] - df_meses['data_pedido']
df_meses['tempo_entrega'] = tempo_demora
df_meses
```

```python
entrega = df_meses.pivot_table(index = 'pais', columns = 'meses', values = 'tempo_entrega',
                              aggfunc = 'mean', sort = False)
entrega
```

```python
estilo_entrega = entrega.style.format('{:,.2f}')



estilo_entrega.set_sticky(axis = 'index')
```

## Hora da prática 5

```python
df_alteracao = df.rename(columns={'pais': 'Custo por país ($)'})
df_clientes = df_alteracao.pivot_table(index = ['tipo_cliente', 'categoria'], columns = 'Custo por país ($)', values = 'custo_produto',
                           aggfunc = 'sum')
df_clientes

```

```python
df_clientes['Total'] = df_clientes.sum(axis = 1)

s_clientes = df_clientes.style.format('{:,.2f}')

tabela = {
    'selector': 'td, th',
    'props': 'font-weight: normal; font-family: Arial; text-align: right; background-color: white'
}
cabecalho = {
    'selector': 'th',
    'props': 'vertical-align: top'
}
s_clientes.set_table_styles([tabela, cabecalho])

s_clientes.set_table_styles({
    ('B2B', 'BlueRay'): [{'selector': 'th', 'props': 'border-top: 1px solid #181818'},
              {'selector': 'td', 'props': 'border-top: 1px solid #181818'}],
    ('B2B', 'Livro'): [{'selector': 'th', 'props': 'font-weight: bold'}],
    ('B2C', 'Livro'): [{'selector': 'th', 'props': 'font-weight: bold'}],
}, overwrite = False, axis = 1)

s_clientes.set_table_styles({
    'Total ($)': [{
        'selector': '.true',
        'props': 'font-weight: bold'
    }]
}, overwrite = False, axis = 0)

cores_coluna = pd.DataFrame(['false','false','false','true','false','false','false','true'],index= df_clientes['Total ($)'].index,
                            columns = ['Total ($)'])

s_clientes.set_td_classes(cores_coluna)
```
